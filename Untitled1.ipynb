factorial

```python
def fact(n):
    if(n==0):
        return 1
    if(n==1):
        return 1
    return n*fact(n-1)
    
        
print(fact(5))
```

```python
def fibbo(n):
    if(n==0 or n==1):
        # print(1,end=" ")
        return 1
    else: 
        return fibbo(n-1)+fibbo(n-2)
    # print(a,end=" ")
    # return a
fibbo(10)
```
